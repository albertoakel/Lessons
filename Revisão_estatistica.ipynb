* Q1. 5-5.01  : P=0.02   (2%)

Observamos que, no intervalo de 0.01h (ou 36 segundos), a chance de uma bact√©ria morrer √© de 2%. Agora, vamos fazer o inverso: aumentar o intervalo. Se 0,01h corresponde a 2%, ent√£o um intervalo 10 vezes maior (0,1h ou 6 minutos) teria uma probabilidade 10 vezes maior, ou seja, 20%. Se aumentarmos ainda mais, para 1h, a probabilidade seria 100 vezes maior que o original, chegando a 200%. Mas espera a√≠... 200%? Como assim? Isso n√£o faz sentido, n√©? Nesses exemplos, a raz√£o entre a probabilidade de morrer durante um intervalo e a dura√ß√£o do intervalo ($\frac{P(\Delta_{t})}{\Delta t} $) √© aproximadamente constante, igual a 2 por hora
(ou $2.hora^{-1}$). Esta quantidade √© chamada de **densidade de probabilidade ou PDF(n√£o √© o arquivo! )**. \
Se quisermos saber a probabilidade de uma bact√©ria morrer em 0.01h (36 segundos), basta multiplicar a densidade de probabilidade pelo intervalo: 2*0.01=0.02 (2%)

**Assim, em resposta √† pergunta qual √© a probabilidade de que a bact√©ria morra em 5 horas?** Usando o que j√° temos, a probabilidade de morrer em um intervalo de tempo $\Delta t$
$$P=2\Delta t = 2 \times 0=zero!$$
Ent√£o a resposta √© zero mesmo üòä

Vamos imaginar um valor pr√≥ximo da hora exata,apertar nossa quest√£o. 5h +1ns, ou seja nosso $\Delta t=10^{-9}s$. Lembrando que estamos trabalhando com hora. Assim:
 $$1s=2.7\times 10^{-4}h$$.
Desta forma, **A probabilidade √© de viver nesse curto espa√ßo de tempo √©: **
$$P=2\Delta t = 2 \times 2.7\times 10^{-4}h=5.4 \times 10^{-13}$$ \

Pensamos em um grupo de bact√©rias e no tempo de vida delas. Se f√¥ssemos agrup√°-las pelo tempo de vida, ter√≠amos uma curva do tempo de vida pela quantidade de bact√©rias. Essa curva, tamb√©m chamada de fun√ß√£o ou distribui√ß√£o, √© uma representa√ß√£o matem√°tica dessa natureza. Aqui, iremos tratar de uma distribui√ß√£o muito espec√≠fica chamada normal, tamb√©m conhecida como fun√ß√£o gaussiana.

---
A equa√ß√£o da distribui√ß√£o normal √© definida por:

$$f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2} \left( \frac{x - \mu}{\sigma} \right)^2}$$

Onde:

$Œº$  m√©dia da distribui√ß√£o; \
$\sigma$ √© o desvio padr√£o; \
$x$ √© a vari√°vel aleat√≥ria;

Algumas caracteristicas;
M√©dia $=$ Mediana($M_{d}$) ( ou quase isso para dados reais)

### Calculando as probabilidades
Em teoria das probabilidades , a fun√ß√£o densidade de probabilidade (FDP) √© uma fun√ß√£o que descreve a verossimilhan√ßa de uma vari√°vel aleat√≥ria tomar um valor dado. A probabilidade da vari√°vel aleat√≥ria cair em uma faixa particular √© dada pela integral da densidade dessa vari√°vel sobre tal faixa - isto √©, √© dada pela √°rea abaixo da fun√ß√£o densidade mas acima do eixo horizontal e entre o menor e o maior valor dessa faixa. Assim.

$$ P[a \leqslant X \leqslant b] =\int_{a}^{b} f(x) dx  $$
Se nossa fun√ß√£o de densidade de probalidade √© conhecidada e √© a gaussina. Ent√£o pode podemos definir a probabilidade entre $x_1$ e $x_2$
$$P[x1:x2]=\int_{x_1}^{x_2} \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2} \left( \frac{x - \mu}{\sigma} \right)^2} dx$$

### **Teste de Hip√≥tese**
### 1. Problema
Suponha que voc√™ tenha uma amostra de notas de alunos em uma prova e queira testar se a m√©dia das notas √© significativamente diferente de 7.0 (um valor de refer√™ncia).

#### 2. Hip√≥tese
*   hip√≥tese nula ($H_{o}$): A m√©dia das notas √© **igual** a 7.0 \
$H_{o}: \mu=7.0$
*   hip√≥tese alternativa ($H_{1}$): A m√©dia das notas √© **diferente** de 7.0 \
$H_{o}: \mu\neq7.0$

#### 3. Dados
supondo a seguinte cole√ß√£o de notas:\
notas=[6.5, 7.2, 6.8, 7.5, 6.9, 7.1, 6.7, 7.4, 6.6, 7.3]

#### 4. M√©todo
Como n, total de elementos da cole√ß√£o √© baixo, podemos usar o m√©todo t-student. Assim definimos a seguinte equa√ß√£o para t:

$$t=\frac{( \bar{x}-\mu_o)\sqrt{\nu}}{s}$$

onde:\
$\bar{x}$ √© a m√©dia da amostra \
$\mu_o$ √© o valor de refer√™ncia \
$s$ √© o desvio da amostra \
$\nu$ graus de liberdada, $\nu=n-1$
###
**passos ** \
calculando a m√©dia, desvio podemos calcular o valor de t, assim: \
$\bar{x}$ = 7.0 \
$s$ = 0.316 \
Finalmente: $t = 0$

**Determina√ß√£o do valor-p** \
O valor-p mede a probabilidade de obter os resultados observados, assumindo que a hip√≥tese nula √© verdadeira. Ele varia de 0 a 1.Em um teste de hip√≥teses, voc√™ formula uma hip√≥tese nula que representa uma afirma√ß√£o a ser testada ‚Äì onde os grupos comparados n√£o tem diferen√ßa entre si ‚Äì, e uma hip√≥tese alternativa ‚Äì que √© a afirma√ß√£o oposta. Indica a probabilidade de obter os resultados observados, assumindo que a hip√≥tese nula seja verdadeira.

* Se o valor-p for menor que o n√≠vel de signific√¢ncia (geralmente $\alpha$ =0,05), rejeitamos $H_o$
* Caso contr√°rio, n√£o rejeitamos $H_o$

O valor-P depende do teste de hip√≥tese proposto.

1.   Teste bicaudal: $H_1:\mu \neq \mu_{o}$ \
      $$valor_p=2P(T>|t|)$$
2.   Teste unicaudal √† direita $H_1:\mu > \mu_{o}$
$$valor_p=P(T>t)$$
3.   Teste unicaudal √† direita $H_1:\mu < \mu_{o}$
$$valor_p=2P(T<t)$$














```python
# Valor de refer√™ncia
valor_referencia = 7.0
notas = [6.5, 7.2, 6.8, 7.5, 6.9, 7.1, 6.7, 7.4, 6.6, 7.3]
# Realizar o teste t para uma amostra
resultado = stats.ttest_1samp(notas, valor_referencia)

# Exibir o resultado
print("Estat√≠stica t:", resultado.statistic)
print("Valor-p:", resultado.pvalue)
resultado
```

```python
import numpy as np

# Fun√ß√£o PDF da distribui√ß√£o normal padr√£o
def pdf_normal(z):
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-z**2 / 2)

# Fun√ß√£o para calcular a integral usando a regra de Simpson
def simpson(f, a, b, n):
    if n % 2 != 0:
        raise ValueError("O n√∫mero de subintervalos (n) deve ser par.")

    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    y = f(x)

    integral = (h / 3) * (y[0] + 4 * np.sum(y[1:-1:2]) + 2 * np.sum(y[2:-2:2]) + y[-1])
    return integral

# Fun√ß√£o para calcular a CDF da distribui√ß√£o normal padr√£o
def cdf_normal(z, n=1000):
    if z < 0:
        return 0.5 - simpson(pdf_normal, z, 0, n)
    else:
        return 0.5 + simpson(pdf_normal, 0, z, n)

# Exemplo de uso
z = 1-0.95  # Valor de Z para calcular a CDF
cdf_z = cdf_normal(z)
print(f"P(Z <= {z}) = {cdf_z:.4f}")
```
